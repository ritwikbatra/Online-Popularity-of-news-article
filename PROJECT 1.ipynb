# Data Extraction and Wrangling from here

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
numerical = ['n_tokens_title', 'n_tokens_content', 'num_hrefs', 'num_self_hrefs', 'num_imgs','num_videos',
            'average_token_length','num_keywords','self_reference_min_shares','self_reference_max_shares',
             'self_reference_avg_sharess']
```

```python
df1=df
df1[numerical]=scaler.fit_transform(df[numerical])
```

```python
df1.head()
```

```python
y11=df1['shares']
df1=df1.drop(['url','timedelta','shares'],axis=1)
```

```python
svd=TruncatedSVD(n_components=10)
arr=svd.fit_transform(df1)
arr=pd.DataFrame(arr)
df12=pd.concat([df1,arr],axis=1)
df12=scaler.fit_transform(df12)
```

```python
x11train,x11test,y11train,y11test=train_test_split(df1,y11,test_size=0.3)
```

```python
predictions=np.zeros((y11test.shape))
```

# MODEL FITTING

```python
bay_rid=BayesianRidge()
bay_rid.fit(x11train,y11train)
bay_pred=bay_rid.predict(x11test)
predictions+=bay_pred
```

```python
lass=Lasso()
lass.fit(x11train,y11train)
lass_pred=lass.predict(x11test)
predictions+=lass_pred
```

```python
ridge=Ridge(alpha=0.5)
ridge.fit(x11train,y11train)
predictions+=ridge.predict(x11test)
```

```python
lin_reg=LinearRegression()
lin_reg.fit(x11train,y11train)
predictions+=lin_reg.predict(x11test)
```

```python
predictions/=4
```

# RESULTS

```python
print("Root mean squared error : %f"%math.sqrt(mean_squared_error(y11test,predictions)))
print("Mean absolute error : %f"%(mean_absolute_error(y11test,predictions)))
```
